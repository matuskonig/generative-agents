```python
def load_data(name: str):
    with open(f"../results/{name}.json", "r") as f:
        result = ExperimentResult.model_validate_json(f.read())

    name_graph: nx.Graph[str] = nx.Graph()
    name_graph.add_edges_from(
        (
            (result.dataset.agents[u].full_name, result.dataset.agents[v].full_name)
            for u, v in result.dataset.edges
        )
    )
    positions = nx.spring_layout(name_graph, seed=42)
    return result, name_graph, positions

A1_baseline_5, A1_baseline_5_graph, A1_baseline_5_positions = load_data("A1_baseline_5")
```

```python
def get_per_epoch_counts(result: ExperimentResult):
    return [
        sum(int(response.final_answer) for response in epoch)
        for epoch in result.epoch_agents_responses
    ]


def get_per_epoch_at_least_once_counts(result: ExperimentResult):
    unique_agents: set[int] = set()
    for epoch in result.epoch_agents_responses:
        epoch_agents = {i for i, response in enumerate(epoch) if response.final_answer}
        unique_agents.update(epoch_agents)
        yield len(unique_agents)

```

```python
def create_abblation_plot(data: list[tuple[str, ExperimentResult]], title: str):
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    JIGGLE_SCALE = 0.025
    max_y = 0
    min_y = 10000

    for (label, result), marker in zip(data, cycle(["o", "s", "^", "D", "v"])):
        jiggle = np.random.uniform(-JIGGLE_SCALE, JIGGLE_SCALE)
        x = np.arange(len(result.epoch_agents_responses)) + 1

        # Subplot 1: per-epoch counts
        y1 = np.array(get_per_epoch_counts(result)) + jiggle
        axes[0].plot(x, y1, label=label, alpha=0.8, marker=marker)
        max_y = max(max_y, max(y1))
        min_y = min(min_y, min(y1))

        # Subplot 2: at-least-once counts
        y2 = np.array(list(get_per_epoch_at_least_once_counts(result))) + jiggle
        axes[1].plot(x, y2, label=label, alpha=0.8, marker=marker)
        max_y = max(max_y, max(y2))
        min_y = min(min_y, min(y2))

    # Common settings
    fig.suptitle(title)
    axes[0].set_xlabel("Epoch")
    axes[1].set_xlabel("Epoch")
    axes[0].set_ylabel("Number of Agents")
    axes[1].set_ylabel("Number of Agents (At Least Once)")
    # axes[1].set_yticks([])  # Hide y-ticks on second subplot for common label

    axes[0].set_ylim(min_y, max_y + 0.5)
    axes[1].set_ylim(min_y, max_y + 0.5)
    plt.legend(loc="upper left", bbox_to_anchor=(0.01, 0.99))

    plt.tight_layout()
    plt.show()


def create_token_consumption_plot(
    data: list[tuple[str, ExperimentResult]],
    title: str,
) -> None:
    """Create a stacked bar chart showing token consumption per experiment."""
    fig, ax = plt.subplots()

    labels = [label for label, _ in data]
    x = np.arange(len(labels))
    width = 0.2

    prompt = [result.prompt_tokens / 1e6 for _, result in data]
    completion = [result.completion_tokens / 1e6 for _, result in data]

    ax.bar(x, prompt, width, label="Prompt Tokens")
    ax.bar(x, completion, width, bottom=prompt, label="Completion Tokens")

    #ax.set_xlabel("Experiment")
    ax.set_ylabel("Tokens (Millions)")
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(labels, rotation=45, ha="right")
    ax.legend()
    ax.grid(axis="y", alpha=0.3)
    plt.tight_layout()
    plt.show()
```
