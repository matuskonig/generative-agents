```python
def get_per_epoch_counts(result: ExperimentResult):
    return [
        sum(int(response.final_answer) for response in epoch)
        for epoch in result.epoch_agents_responses
    ]


def get_per_epoch_at_least_once_counts(result: ExperimentResult):
    unique_agents: set[int] = set()
    for epoch in result.epoch_agents_responses:
        epoch_agents = {i for i, response in enumerate(epoch) if response.final_answer}
        unique_agents.update(epoch_agents)
        yield len(unique_agents)

```

```python
def create_abblation_plot(data: list[tuple[str, ExperimentResult]], title: str, filename: str) -> None:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    JIGGLE_SCALE = 0.025
    max_y = 0
    min_y = 10000

    for (label, result), marker in zip(data, cycle(["o", "s", "^", "D", "v"])):
        jiggle = np.random.uniform(-JIGGLE_SCALE, JIGGLE_SCALE)
        x = np.arange(len(result.epoch_agents_responses)) + 1

        # Subplot 1: per-epoch counts
        y1 = np.array(get_per_epoch_counts(result)) + jiggle
        axes[0].plot(x, y1, label=label, alpha=0.8, marker=marker)
        max_y = max(max_y, max(y1))
        min_y = min(min_y, min(y1))

        # Subplot 2: at-least-once counts
        y2 = np.array(list(get_per_epoch_at_least_once_counts(result))) + jiggle
        axes[1].plot(x, y2, label=label, alpha=0.8, marker=marker)
        max_y = max(max_y, max(y2))
        min_y = min(min_y, min(y2))

    # Common settings
    #fig.suptitle(title)
    axes[0].set_xlabel("Epoch")
    axes[1].set_xlabel("Epoch")
    axes[0].set_ylabel("Number of Agents")
    axes[1].set_ylabel("Number of Agents (At Least Once)")
    # axes[1].set_yticks([])  # Hide y-ticks on second subplot for common label

    axes[0].set_ylim(min_y, max_y + 0.5)
    axes[1].set_ylim(min_y, max_y + 0.5)
    plt.legend(loc="upper left", bbox_to_anchor=(0.01, 0.99))

    plt.tight_layout()
    plt.savefig(f"assets/{filename}", bbox_inches="tight")
    plt.show()


def create_token_consumption_plot(
    data: list[tuple[str, ExperimentResult]],
    title: str,
    filename: str,
) -> None:
    """Create a stacked bar chart showing token consumption per experiment."""
    fig, ax = plt.subplots()

    labels = [label for label, _ in data]
    x = np.arange(len(labels))
    width = 0.2

    prompt = [result.prompt_tokens / 1e6 for _, result in data]
    completion = [result.completion_tokens / 1e6 for _, result in data]

    ax.bar(x, prompt, width, label="Prompt Tokens")
    ax.bar(x, completion, width, bottom=prompt, label="Completion Tokens")

    #ax.set_xlabel("Experiment")
    ax.set_ylabel("Tokens (Millions)")
    #ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(labels, rotation=45, ha="right")
    ax.legend()
    ax.grid(axis="y", alpha=0.3)
    plt.tight_layout()
    plt.savefig(f"assets/{filename}", bbox_inches="tight")
    plt.show()
```

# Introduction

In this abblation study, we will compare our baseline solution across various different setups and find a most powerful combination
of hyperparameters to get the optimal solution. In addition, this study aims to describe the general behavior of the agents.

The baseline solution is as follows:
- embedding memory with mean std (0.5) selector
- classical conversational updates
- forgetting with a 50% max linear probability
- information spread selector
- BDI for planning
- 10 epochs, 10 utterances per conversation

The abblation study will investigate every individual component of this solution to come up with the best combination of hyperparameters.

In this setup, we tested the robustness of the systems by changing the prompts of the system. We removed any mention of expected behavior,
in terms of information sharing, conversation and general expected behavoior of artificial agent. 

We tested it together with removing the BDI planning layer and we can see the results above.

As possibly expected, the setup without nor BDI nor proper prompts has worse results. In 5 agent setup, the number of agents stays flat both in per-epoch and cumulative statistics. In 10-agent setup, the cumulative statistics converged to the common value, however the per-epoch values are worse than the standard variants.

In 5 agent setup the baseline setup is better in all statistics. In 10 agent setup the baseline setup has faster momentum, however they still converged to the same cumulative statistics. In addition, the per-epoch statistics declined, which suggests the agents are more prone to forgetting.

From this scenario we assume the baseline might be better that no-BDI setup. This experiments also confirms our prompts are designed well,
however it is worrying that the behavior is prone to prompt design.

Regarding the token consumption, looks like the BDI is adding 20% more compute on top. In addition, reducing the prompts also reduce the number of tokens burns.

Unitary notes is an alternative mechanism to classical memory update mechanism.
In classical conversational memory, the agent appends the newly learned information into the memory. The number of added information is not limited and additional pruning and selection mechanism is needed to keep the scope managable for the LLM to process.
Unitary notes however reduce this by having the agent store and update just a single note for any other agent it encounters, limiting the size of the memories.

The results are similar in a way. Unitary notes offers a slightly worse performance, however the characteristics of both baseline solution and unitary notes are really similar. On the positive note, unitary notes have a much better compute cost, where baseline solution is 20% more expensive.

While Top-K is conservative strategy, the variable strategies are designed with statistics in mind, where they are inspired by how the statistical significance works. 

In the 5 agent setup the Top-K selection performed better than any other selection. Mean-std selection however had the steepest increase. Top-std is lacking.

In 10 agent setup the situation is fairly different. Top K still scored the best results, however top k (25) was much better in comparison to other methods. Mean-std are lacking, while top-std are performing just a little worse than Top-K.

In the baseline solution, forgetting mechanism is enabled. That mechanism works by assigning a probability to each memory record linearly to the age of that memory. In the baseline solution, the maximal probability is 50 %, that means the probability is rescaled and thus linearly increasing to 50 % for the oldest record.

In this experiment, we compared it to the 100% forgetting rule and no forgetting at all.

# Results, discussion and next steps

## Summary of Key Findings

### Component Impact Ranking

| Component              | Impact on Performance    | Impact on Cost | Recommendation   |
| ---------------------- | ------------------------ | -------------- | ---------------- |
| **BDI Planning**       | High (critical)          | +20% tokens    | **Keep**         |
| **Embedding Memory**   | Medium (scale-dependent) | Moderate       | **Keep**         |
| **Retrieval Strategy** | High                     | Correlated     | Use Top-K (k=25) |
| **Forgetting**         | Low                      | None           | 50% max default  |
| **Selector**           | Low-Medium               | Moderate       | Full parallel    |
| **Prompts**            | Medium                   | Low            | Keep current     |

### Critical Observations

1. **BDI is the most impactful component** - removing it causes significant decline in both 5 and 10 agent setups
2. **Embedding memory becomes critical at scale** - simple memory might struggle with context overload
3. **Top-K retrieval outperforms statistical methods** - simpler is better for this task, although results not sharply convincing
4. **Information spread selector underperforms** - full parallel is slightly better despite being less principled, however theoretically they are equivalent
5. **System is prompt-sensitive** - behavior changes with prompt wording (concerning for robustness)
6. **E10U10 is minimum viable** - less compute yields incomplete saturation

### Limitations of Current Study

- No statistical tests performed - all conclusions are visual/qualitative
- Single run per configuration - high variance in LLM outputs not captured
- Limited scale (max 10 agents) - behavior at larger scales unknown

## Statistical Analysis Plan

### Primary Metric

**Final cumulative knowledge** — number of agents who knew the information at least once by the final epoch.

### Possible statistical Tests

| Test                        | Purpose                              | When to Use                       |
|-----------------------------|--------------------------------------|-----------------------------------|
| Wilcoxon signed-rank        | Compare paired runs (matched seeds)  | Default (non-parametric)          |
| Paired t-test               | Compare paired runs                  | If Shapiro-Wilk p > 0.05          |
| Effect size (Cohen's d)     | Practical significance               | Always report alongside p-values  |
| Bootstrap CI                | Confidence intervals                 | For robustness                    |

### Secondary Metrics

1. **Area Under Curve (AUC)** — cumulative knowledge integrated over all epochs
2. **Convergence epoch** — epoch at which 50% of agents first knew the information
3. **Per-epoch stability** — variance in per-epoch counts (lower = more stable recall)
4. **Token efficiency** — knowledge gained per million tokens
